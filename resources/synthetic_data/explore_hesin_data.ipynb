```python
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
hospital = pd.read_csv('hesin.csv')
hospital.columns
```

```python
hospital_rem_dupl = hospital.drop_duplicates(subset=['eid', 'ins_index', 'dsource'], keep='first')
hospital_rem_dupl
```

```python
hospital_rem_dupl['hosp_dupl'] = hospital_rem_dupl.duplicated(subset=['eid', 'ins_index'], keep=False)
hospital_dupl = hospital_rem_dupl[hospital_rem_dupl['hosp_dupl']==True].sort_values(['eid', 'ins_index'])
hospital_dupl[['eid', 'ins_index', 'dsource', 'admidate', 'disdate']]
```

```python
for i in hospital_dupl['eid']:
    dupl = hospital_dupl[['eid', 'ins_index', 'admidate', 'disdate']][hospital_dupl['eid']==i]
    print(dupl)
```
